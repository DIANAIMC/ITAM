```r
# obtiene regesión lineal
# Hace lo mismo que arriba
lm.fit <- lm(Peso ~ Altura,
             data = df)
lm.fit
```

Este resultado muestra el resultado de un modelo de regresión lineal que se ajusta a los datos en el data frame "df". La relación entre las variables "Peso" y "Altura" se ha modelado mediante la fórmula "Peso ~ Altura".

Los residuos son las diferencias entre los valores reales y los valores predichos por el modelo. Los residuos tienen un rango de -23.5567 a 21.2480.

Los coeficientes muestran el valor estimado de los parámetros del modelo, que son la pendiente (137.8151) y el intercepto (-159.0917) de la línea de regresión. También se proporciona la desviación estándar del error (Std. Error) y el valor "t" asociado a cada coeficiente, que se utiliza para determinar si es significativo estadísticamente. En este caso, ambos coeficientes tienen un valor p muy pequeño (menor que 2e-16), lo que indica que son significativos estadísticamente.

El "Residual standard error" es la desviación estándar de los residuos, y en este caso es de 5.543. El "R-squared" es una medida de qué tan bien se ajusta el modelo a los datos y en este caso es de 0.8552, lo que indica que el modelo explica un 85.52% de la varianza en los datos. El "F-statistic" y su p-value también proporcionan información sobre la significancia global del modelo, y en este caso, el p-value es muy pequeño, lo que indica que el modelo es significativo estadísticamente.

Esto quiere decir que nuestra predicción varía con nuestros datos 5 kilos aprox

```r
# calcula error cuadrado medio y la raiz del error cuadrado medio
errors <- residuals(lm.fit) 
squared.errors <- errors ^ 2
mse <- mean(squared.errors)
rmse <- sqrt(mse)
rmse
```

```r
# Incluye bibliotéca de gráficas
library('ggplot2')
```

```r
# lee datos separados por tabuladores
top.1000.sites <- read.csv('top_1000_sites.tsv',
                           sep = '\t',
                           stringsAsFactors = FALSE)
```

```r
# Grafica datos como dispersión de puntos
ggplot(top.1000.sites, aes(y = PageViews, x = UniqueVisitors)) +
  geom_point()
```

```r
# Grafica distribución de los datos
ggplot(top.1000.sites, aes(x = PageViews)) +
  geom_density()
```

```r
# Grafica distribución del logaritmo de los datos
ggplot(top.1000.sites, aes(x = log(PageViews))) +
  geom_density()  
```

```r
# Grafíca logaritmo de los datos como dispersión de puntos
ggplot(top.1000.sites, aes(y = log(PageViews), x = log(UniqueVisitors))) +
  geom_point()
```

```r
# Grafíca logaritmo de los datos como dispersión de puntos más linea de regresión
ggplot(top.1000.sites, aes(y = log(PageViews), x = log(UniqueVisitors))) +
  geom_point() +
  geom_smooth(method = 'lm', se = FALSE, formula = 'y ~ x')
```

```r
# obtiene regesión lineal
lm.fit <- lm(log(PageViews) ~ log(UniqueVisitors) ,
             data = top.1000.sites)

# calcula error cuadrado medio y la raiz del error cuadrado medio
errors <- residuals(lm.fit) 
squared.errors <- errors ^ 2
mse <- mean(squared.errors)
rmse <- sqrt(mse)
rmse

# calcula r cuadrada
r2 <- 1-sum(squared.errors)/sum((log(top.1000.sites$PageViews)-mean(log(top.1000.sites$PageViews)))^2)
r2
```

```r
# agrega variables categóricas
lm.fit <- lm(log(PageViews) ~ HasAdvertising + log(UniqueVisitors) + InEnglish,
             data = top.1000.sites)

#calcula índices
errors <- residuals(lm.fit) 
squared.errors <- errors ^ 2
mse <- mean(squared.errors)
rmse <- sqrt(mse)
rmse

r2 <- 1-sum(squared.errors)/sum((log(top.1000.sites$PageViews)-mean(log(top.1000.sites$PageViews)))^2)
r2
```

```r
# Regresión lineal sobre función cuadrática

set.seed(1)

x <- seq(-10, 10, by = 0.01)
y <- 1 - x ^ 2 + rnorm(length(x), 0, 5)

ggplot(data.frame(X = x, Y = y), aes(x = X, y = Y)) + 
  geom_point() +
  geom_smooth(method = 'lm', se = FALSE, formula = 'y ~ x')
```

```r
# transformación de entradas para convertir el problema en lineal

x.squared <- x ^ 2

ggplot(data.frame(XSquared = x.squared, Y = y), aes(x = XSquared, y = Y)) +
  geom_point() +
  geom_smooth(method = 'lm', se = FALSE, formula = 'y ~ x')
```

```r
summary(lm(y ~ x))$r.squared
```

```r
summary(lm(y ~ x.squared))$r.squared
```

```r
# Polinomios

set.seed(1)

x <- seq(0, 1, by = 0.01)
y <- sin(2 * pi * x) + rnorm(length(x), 0, 0.1)
df <- data.frame(X = x, Y = y)

ggplot(df, aes(x = X, y = Y)) + geom_point()

summary(lm(Y ~ X, data = df))$r.squared
```

```r
ggplot(data.frame(X = x, Y = y), aes(x = X, y = Y)) + geom_point() +
  geom_smooth(method = 'lm', se = FALSE, formula = 'y ~ x')
```

```r
library(dplyr)
```

```r
df <- mutate(df, X2 = X ^ 2) 

df <- df %>% mutate(X3 = X ^ 3)

summary(lm(Y ~ X + X2 + X3, data = df))$r.squared
```

```r
# Polinómios ortogonales

poly.fit <- lm(Y ~ poly(X, degree = 1), data = df) 
df <- mutate(df, PredictedY = predict(poly.fit))
ggplot(df, aes(x = X, y = PredictedY)) + geom_point() +
  geom_line()
```

```r
poly.fit <- lm(Y ~ poly(X, degree = 3), data = df) 
df <- mutate(df, PredictedY = predict(poly.fit))
ggplot(df, aes(x = X, y = PredictedY)) + geom_point() +
  geom_line()
```

```r
poly.fit <- lm(Y ~ poly(X, degree = 5), data = df) 
df <- mutate(df, PredictedY = predict(poly.fit))
ggplot(df, aes(x = X, y = PredictedY)) + geom_point() +
  geom_line()
```
