# Good linear relationship

```r
x <- rnorm(50, mean=15, sd=3)
slope <- 3.2
intercept <- 7.5
errors <- rnorm(50, sd=2)
y <- slope * x + intercept + errors
reg <- lm(y~x)
plot(x,y,pch=16, las=1)
abline(reg, lwd=2, col="blue")
par(mar = c(2, 2, 2, 2))
par(mfrow=c(2,2))
plot(reg)
```

# Parábola

```r
x <- rnorm(50, mean=-2, sd=3)
quadratic <- 0.5
slope <- 3.2
intercept <- 7.5
errors <- rnorm(50, sd=5)
y <- quadratic*x*x + slope*x + intercept + errors
reg <- lm(y~x)

plot(x,y,pch=16, las=1)
abline(reg, lwd=2, col="blue")
par(mar = c(2, 2, 2, 2))
par(mfrow=c(2,2))
plot(reg)
```

# Increasing variance

```r
x <- rnorm(50, mean=15, sd=3)
slope <- 3.2
intercept <- 7.5
errors <- rnorm(50,sd=2)
y <- slope * x + intercept + errors*((x-min(x))/2)
reg <- lm(y~x)

plot(x,y,pch=16, las=1)
abline(reg, lwd=2, col="blue")
par(mar = c(2, 2, 2, 2))
par(mfrow=c(2,2))
plot(reg)
```

# Outlier

```r
x <- rnorm(50, mean=15, sd=3)
slope <-  3.2
intercept <-  7.5
y <- slope * x + intercept + errors # good line
x <- c(x,50) # adding an outlier
y <- c(y,30)
reg <- lm(y~x)

plot(x,y,pch=16, las=1)
abline(reg, lwd=2, col="blue")
par(mar = c(2, 2, 2, 2))
par(mfrow=c(2,2))
plot(reg)
```
